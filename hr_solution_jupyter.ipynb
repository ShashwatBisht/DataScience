```python
import pymysql
import pandas as pd

conn = pymysql.connect(
    host='localhost',  
    user='root', 
    password='shashwat12', 
    database='hr'
)

cursor = conn.cursor()
```

```python
def execute_query(query):
    cursor = conn.cursor()
    cursor.execute(query)

    columns = [desc[0] for desc in cursor.description]
    data = cursor.fetchall()

    df = pd.DataFrame(data, columns=columns)

    cursor.close()
    return df  
```

# 1. Retrieve all details of employees

```python
query1 = "SELECT * FROM employees;"
execute_query(query1)
```

# 2. Display first name, last name, and email of all employees

```python
query2 = "SELECT first_name, last_name, email FROM employees;"
execute_query(query2)
```

# 3. Retrieve distinct job titles from the jobs table


```python
query3 = "SELECT DISTINCT job_title FROM jobs;"
execute_query(query3)
```

# 4. Find the total number of employees in the company


```python
query4 = "SELECT COUNT(*) AS total_employees FROM employees;"
execute_query(query4)
```

# 5. Retrieve employees hired after January 1, 2015


```python
query5 = "SELECT * FROM employees WHERE hire_date > '2015-01-01';"
execute_query(query5)
```

# 6. List all employees who have a salary greater than 5000


```python
query6 = "SELECT * FROM employees WHERE salary > 5000;"
execute_query(query6)
```

# 7. Retrieve employees with job titles containing the word 'Manager'


```python
query7 = "SELECT * FROM jobs WHERE job_title LIKE '%Manager%';"
execute_query(query7)
```

# 8. Retrieve all employees whose first name starts with 'A' and ends with 'n'


```python
query8 = "SELECT * FROM employees WHERE first_name LIKE 'A%n';"
execute_query(query8)
```

# 9. Display employees who do not have a commission


```python
query9 = "SELECT * FROM employees WHERE commission_pct IS NULL;"
execute_query(query9)
```

# 10. Retrieve the top 5 highest-paid employees


```python
query10 = "SELECT * FROM employees ORDER BY salary DESC LIMIT 5;"
execute_query(query10)
```

# 11. Find the average salary of all employees


```python
query11 = "SELECT AVG(salary) AS average_salary FROM employees;"
execute_query(query11)
```

# 12. Retrieve the total number of employees working in each department


```python
query12 = "SELECT department_id, COUNT(*) AS num_employees FROM employees GROUP BY department_id;"
execute_query(query12)
```

# 13. Display the employee's first name and the length of their first name


```python
query13 = "SELECT first_name, LENGTH(first_name) AS name_length FROM employees;"
execute_query(query13)
```

# 14. Convert the hire_date of employees to display only the year


```python
query14 = "SELECT first_name, last_name, YEAR(hire_date) AS hire_year FROM employees;"
execute_query(query14)
```

# 15. Retrieve the minimum and maximum salary for each job title


```python
query15 = "SELECT job_id, MIN(salary) AS min_salary, MAX(salary) AS max_salary FROM employees GROUP BY job_id;"
execute_query(query15)
```

# 16. Retrieve employee names along with their department names


```python
query16 = """
SELECT e.first_name, e.last_name, d.department_name
FROM employees e
JOIN departments d ON e.department_id = d.department_id;
"""
execute_query(query16)
```

# 17. List employees along with their job titles and department location


```python
query17 = """
SELECT e.first_name, e.last_name, j.job_title, l.city, l.state_province
FROM employees e
JOIN jobs j ON e.job_id = j.job_id
JOIN departments d ON e.department_id = d.department_id
JOIN locations l ON d.location_id = l.location_id;
"""
execute_query(query17)
```

# 18. Retrieve department names along with the count of employees in each department


```python
query18 = """
SELECT d.department_name, COUNT(e.employee_id) AS num_employees
FROM departments d
LEFT JOIN employees e ON d.department_id = e.department_id
GROUP BY d.department_name;
"""
execute_query(query18)
```

# 19. Find employees who have the same job as their manager


```python
query19 = """
SELECT e1.first_name, e1.last_name
FROM employees e1
JOIN employees e2 ON e1.manager_id = e2.employee_id
WHERE e1.job_id = e2.job_id;
"""
execute_query(query19)
```

# 20. Display names of employees who worked in different jobs in the past (job_history)


```python
query20 = """
SELECT DISTINCT e.first_name, e.last_name
FROM employees e
JOIN job_history jh ON e.employee_id = jh.employee_id;
"""
execute_query(query20)
```
